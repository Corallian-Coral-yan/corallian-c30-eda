```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from PIL import Image

np.random.seed(1)
```

```python
raw_df = pd.read_csv("c30-filename-index.csv")
complete_df = raw_df.fillna("")
```

```python
index_sample = complete_df.sample(9)
index_sample
```

```python

def display_images(filepaths, r=3, c=3):
    fig = plt.figure(figsize=(10, 8))

    for i, filepath in enumerate(filepaths):
        if filepath:
            img = Image.open(filepath, "r")
            img_np = np.array(img)
            ax = fig.add_subplot(r, c, i + 1)
            ax.set_axis_off()

            plt.imshow(img_np)
```

```python
display_images(index_sample["raw_image_filepath"])
```

```python
display_images(index_sample["sorted_image_filepath"])
```

```python
display_images(index_sample["scored_image_filepath"])
```
